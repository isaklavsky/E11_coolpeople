```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = "pm25.csv"  # Ensure the file is in the same directory as this script
df = pd.read_csv(file_path)
```

```python
# Print column names to debug issues
print("Columns in dataset:", df.columns)
```

```python
for col in df.columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')

plt.figure(figsize=(10, 6))
```

```python
similar_cols = [col for col in df.columns if "PM" in col]
for col in similar_cols:
    plt.hist(df[col].dropna(), bins=20, alpha=0.6, label=col.replace("\t", " "), edgecolor='black')

plt.xlabel("Concentration (µg/m³)")
plt.ylabel("Frequency")
plt.title("Frequency Distribution of PM Data")
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Separate histograms for other particle counts
for col in df.columns:
    if col not in similar_cols and col != "Time":
        plt.figure(figsize=(8, 5))
        plt.hist(df[col].dropna(), bins=20, edgecolor='black', alpha=0.7)
        plt.xlabel("Count")
        plt.ylabel("Frequency")
        plt.title(f"Frequency Distribution of {col.replace('\t', ' ')}")
        plt.grid(axis='y', linestyle='--', alpha=0.7)
        plt.show()
```
