```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Loc1_session1_group1_2025-04-16_16-05-24.csv")
```

```python
# Convert time to datetime
# the time is in different time zone (standard time)
df['Time'] = pd.to_datetime(df['Time'], unit='s')
```

```python
# Part a: Radiation Counts
plt.figure(figsize=(12, 4))
plt.plot(df['Time'], df['counts'], label='Radiation Counts')
plt.xlabel('Time')
plt.ylabel('Counts')
plt.title('Radiation Counts over Time')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Part a: Temperature
plt.figure(figsize=(12, 4))
plt.plot(df['Time'], df['Temperature'], label='Temperature (C)', color='orange')
plt.xlabel('Time')
plt.ylabel('Temperature (C)')
plt.title('Temperature over Time')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Plot PM2.5
plt.figure(figsize=(12, 4))
plt.plot(df['Time'], df['PM25'], label='PM2.5 (g/m)', color='green')
plt.xlabel('Time')
plt.ylabel('PM2.5 (g/m)')
plt.title('PM2.5 Concentration over Time')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Part b: Plot histograms and compute mean + standard deviation
variables = ['counts', 'Temperature', 'PM25']

for var in variables:
    mean_val = df[var].mean()
    std_val = df[var].std()
    print(f"{var} - Mean: {mean_val:.2f}, Standard Deviation: {std_val:.2f}")
    
    plt.figure(figsize=(8, 4))
    plt.hist(df[var], bins=30, color='skyblue', edgecolor='black')
    plt.axvline(mean_val, color='red', linestyle='dashed', linewidth=1.5, label=f'Mean = {mean_val:.2f}')
    plt.axvline(mean_val + std_val, color='green', linestyle='dotted', linewidth=1.5, label=f'+1 = {mean_val + std_val:.2f}')
    plt.axvline(mean_val - std_val, color='green', linestyle='dotted', linewidth=1.5, label=f'-1 = {mean_val - std_val:.2f}')
    plt.title(f'{var} Distribution')
    plt.xlabel(var)
    plt.ylabel('Frequency')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

The distributions of counts and PM2.5 show a somewhat vague bell shape but still deviate from a true normal distribution due to noticeable skew and irregular peaks. Temperature, on the other hand, clearly doesn't look normally distributed, likely due to moving between different environmental zones. Overall, these deviations are expected in mobile data collection, where local sources and environmental changes introduce variability.

#### Part c:

Based on statistics calculated above, the expected variation in radiation counts would be approximately $\sqrt{29.05} \approx 5.39$ using the mean. However, the actual standard deviation measured/calculated from the data is 4.38, which is slightly lower than expected. Suggesting that the variability in the data is somewhat less than what pure counting statistics would predict, possibly due to averaging effects, limited mobility during measurement, or some smoothing in the measurement process.

```python
import numpy as np

# Part d: Compare expected vs. measured standard deviation for counts

# Counts data
mean_counts = df['counts'].mean()
std_measured = df['counts'].std()
n_samples = len(df)

# Expected standard deviation from counting statistics
std_expected = np.sqrt(mean_counts)

# Uncertainty on measured std dev using CLT
uncertainty_std = std_expected / np.sqrt(2 * np.sqrt(n_samples))

print("Part d:")
print(f"Mean Counts (N): {mean_counts:.2f}")
print(f"Expected Std Dev (N): {std_expected:.2f}")
print(f"Measured Std Dev: {std_measured:.2f}")
print(f"Uncertainty on Measured Std Dev: {uncertainty_std:.2f}")

if abs(std_measured - std_expected) <= uncertainty_std:
    print(" The measured and expected standard deviations are in statistical agreement.")
else:
    print(" The measured and expected standard deviations are NOT in statistical agreement.")
```

Using the CLT, the uncertainty on the measured standard deviation is approximately 1.16. The expected standard deviation (N) is 5.39, and the measured value is 4.38. Since the difference is within the estimated uncertainty ($5.39-4.38 = 1.01$), the variation in counts is consistent with what wed expect from counting statistics.

#### Part e:

The measured variation in radiation counts was less than or approximately equal to the expected variation from simple counting statistics, within the estimated uncertainty. Meaning that the observed radiation levels are consistent with what wed expect from statistical fluctuations in counting. If the measured variation had been greater, it might have implied additional sources of variability such as nearby radiation-emitting objects, changes in shielding, or environmental conditions. But since the variation is statistically consistent, the dominant source of randomness appears to be the inherent nature of radioactive decay itself.

```python
# Part f:
# Round latitude and longitude
df['LatLng'] = df['Latitude'].round(5).astype(str) + "," + df['Longitude'].round(5).astype(str)

# Group
location_stats = df.groupby('LatLng')['counts'].agg(['mean', 'std', 'count']).reset_index()
location_stats['sem'] = location_stats['std'] / np.sqrt(location_stats['count'])

# Display locations and stats
print("Part f: Radiation by Location")
for _, row in location_stats.iterrows():
    print(f"Location {row['LatLng']}: Mean = {row['mean']:.2f}, SEM = {row['sem']:.2f}")
```

There are small differences in the mean radiation counts between different locations, and in some cases, these differences are larger than the uncertainty, suggesting statistical significance. But even if statistically significant, the differences in radiation levels are relatively small in absolute terms (a few counts per second). These variations correspond to very low radiation dose differences, which are not biologically or health-significant. They remain well below typical background radiation levels and pose no meaningful health risk.

```python
# Sort locations by mean counts for readability
location_stats_sorted = location_stats.sort_values(by='mean', ascending=False)

plt.figure(figsize=(14, 6))
plt.bar(location_stats_sorted['LatLng'], location_stats_sorted['mean'], yerr=location_stats_sorted['sem'],
        capsize=4, color='skyblue', edgecolor='black')
plt.xticks(rotation=90, fontsize=8)
plt.xlabel('Location')
plt.ylabel('Mean Radiation Counts')
plt.title('Mean Radiation Counts by Location')
plt.tight_layout()
plt.grid(axis='y')
plt.show()
```
