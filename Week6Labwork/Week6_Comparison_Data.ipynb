```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
inside_data = pd.read_csv("inside.csv")
display(inside_data)
outside_data = pd.read_csv("outside.csv")
display(outside_data)
inside_data2 = pd.read_csv("Liams_inside.csv")
display(inside_data2)
outside_data2 = pd.read_csv("Liams_outside.csv")
display(outside_data2)
```

```python
print("Inside CSV Columns:", inside_data2.columns.tolist())
print("Outside CSV Columns:", outside_data2.columns.tolist())
```

```python
columns_to_drop = [
    "Particles > 2.5um / 0.1L air",
    "Particles > 5.0um / 0.1L air",
    "Particles > 10um / 0.1L air",
    'Particles > 0.3um / 0.1L air', 
    'Particles > 0.5um / 0.1L air', 
    'Particles > 1.0um / 0.1L air'
]
inside_data = inside_data.drop(columns=[col for col in columns_to_drop if col in inside_data.columns])
outside_data = outside_data.drop(columns=[col for col in columns_to_drop if col in outside_data.columns])
```

```python
columns_to_drop = [
    "pressure",
    "altitude"
]
inside_data2 = inside_data2.drop(columns=[col for col in columns_to_drop if col in inside_data2.columns])
outside_data2 = outside_data2.drop(columns=[col for col in columns_to_drop if col in outside_data2.columns])
```

```python
def timenormalize(dataframe1):
    # subtracts first value of first column in data frame against all other values in first column
    # intention is to make time readable by removing epoch time difference
    dataframe1.iloc[:,0]-=dataframe1.iloc[0,0]
    # print(f"This is dataframetest: {dataframe1}")

timenormalize(inside_data)
timenormalize(outside_data)
timenormalize(inside_data2)
timenormalize(outside_data2)
```

```python
# Plot time series for each variable
variables = ["Temp (C)", "Gas (Ohm)", "Humidity (%)", 'PM 1.0: %d',
             'PM 2.5: %d','PM 10: %d']

for var in variables:
    plt.figure(figsize=(10, 5))
    plt.plot(inside_data["Time"], inside_data[var], label="Inside", linestyle="dashed")
    plt.plot(inside_data2["Time"], inside_data2[var], label="Inside 2")
    plt.xlabel("Time (s)")
    plt.ylabel(var)
    plt.title(f"Time Series of {var}")
    plt.legend()
    plt.grid()
    plt.show()
```

```python
# Plot time series for each variable
for var in variables:
    plt.figure(figsize=(10, 5))
    plt.plot(outside_data["Time"], outside_data[var], label="Outside", linestyle="dashed")
    plt.plot(outside_data2["Time"], outside_data2[var], label="Outside 2")
    plt.xlabel("Time (s)")
    plt.ylabel(var)
    plt.title(f"Time Series of {var}")
    plt.legend()
    plt.grid()
    plt.show()
```

```python
# histograms
for var in variables:
    plt.figure(figsize=(8, 5))
    plt.hist(inside_data[var], bins=20, alpha=0.5, label="Inside", density=True)
    plt.hist(inside_data2[var], bins=20, alpha=0.5, label="Inside 2", density=True)
    plt.xlabel(var)
    plt.ylabel("Density")
    plt.title(f"Histogram of {var}")
    plt.legend()
    plt.grid()
    plt.show()
```

```python
# histograms 2
for var in variables:
    plt.figure(figsize=(8, 5))
    plt.hist(outside_data[var], bins=20, alpha=0.5, label="Outside", density=True)
    plt.hist(outside_data2[var], bins=20, alpha=0.5, label="Outside 2", density=True)
    plt.xlabel(var)
    plt.ylabel("Density")
    plt.title(f"Histogram of {var}")
    plt.legend()
    plt.grid()
    plt.show()
```

```python
# Comparison
for var in variables:
    mean_inside = np.mean(inside_data[var])
    mean_inside2 = np.mean(inside_data2[var])
    std_inside = np.std(inside_data[var], ddof=1)
    std_inside2 = np.std(inside_data2[var], ddof=1)
    
    N_inside = len(inside_data[var])
    N_inside2 = len(inside_data2[var])
    
    sigma_mean_inside = std_inside / np.sqrt(N_inside)
    sigma_mean_inside2 = std_inside2 / np.sqrt(N_inside2)
    
    diff = abs(mean_inside - mean_inside2)
    threshold = 3 * np.sqrt(sigma_mean_inside**2 + sigma_mean_inside2**2)
    
    print(f"{var}:")
    print(f"  Inside Mean ± σ_mean: {mean_inside:.2f} ± {sigma_mean_inside:.2f}")
    print(f"  Inside 2 Mean ± σ_mean: {mean_inside2:.2f} ± {sigma_mean_inside2:.2f}")
    print(f"  Difference = {diff:.2f}, Threshold = {threshold:.2f}")
    
    if diff > threshold:
        print("Statistically different (3𝜎 separated)")
    else:
        print("Not statistically different")
    print("-" * 50)
```
