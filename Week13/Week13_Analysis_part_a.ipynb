```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Loc1_session1_group1_2025-04-16_16-05-24.csv")
```

```python
# Convert time to datetime
df['Time'] = pd.to_datetime(df['Time'], unit='s')
```

```python
# Plot Radiation Counts
plt.figure(figsize=(12, 4))
plt.plot(df['Time'], df['counts'], label='Radiation Counts')
plt.xlabel('Time')
plt.ylabel('Counts')
plt.title('Radiation Counts over Time')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Plot Temperature
plt.figure(figsize=(12, 4))
plt.plot(df['Time'], df['Temperature'], label='Temperature (C)', color='orange')
plt.xlabel('Time')
plt.ylabel('Temperature (C)')
plt.title('Temperature over Time')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Plot PM2.5
plt.figure(figsize=(12, 4))
plt.plot(df['Time'], df['PM25'], label='PM2.5 (g/m)', color='green')
plt.xlabel('Time')
plt.ylabel('PM2.5 (g/m)')
plt.title('PM2.5 Concentration over Time')
plt.grid(True)
plt.tight_layout()
plt.show()
```
